```
#Easy mode: Using Pandas or any SQL (python library), find the strongest Pokemon by comparing their stats. You can use this formula:([Attack of attacking pokemon] - [Defense of defending pokemon])*0.8 + ([Sp. Attack of attacking pokemon] - [Sp. Defense of defending pokemon])*0.2
#formula:([Attack of attacking pokemon] - [Defense of defending pokemon])*0.8 + ([Sp. Attack of attacking pokemon] - [Sp. Defense of defending pokemon])*0.2
formula=(df.ATK-df.DEF)*0.8+(df.SP_ATK-df.SP_DEF)*0.2
df['Formula'] = formula
#display(df)
sorted_df = df.sort_values(by='Formula',ascending=False)
display(sorted_df)

```
